```python
for i in range(len(extracted['statutes_split'])):
    for j in range(len(extracted['statutes_split'][i])):
        statute = extracted['statutes_split'][i][j]
        statute_low = statute.lower()
        if 'food' in statute_low or 'fss' in statute_low:
            extracted['statutes_split'][i][j] = 'Food Safety and Standards Act, 2006'
        if 'cotp' in statute_low or 'cigarette' in statute_low:
            extracted['statutes_split'][i][j] = 'Cigarettes and Other Tobacco Products Act, 2003'
        if 'ipc' in statute_low or 'penal code' in statute_low:
            extracted['statutes_split'][i][j] = 'Indian Penal Code'
        if 'cgst' in statute_low or 'gst' in statute_low or 'goods and services tax' in statute_low:
            extracted['statutes_split'][i][j] = 'Central Goods and Services Tax Act, 2017'
```

```python
# remove years from statutes

for i in range(len(extracted['statutes_split'])):
    for j in range(len(extracted['statutes_split'][i])):
        statute = extracted['statutes_split'][i][j]
        extracted['statutes_split'][i][j] = re.sub(r'\d{4}', '', statute)
        #remove commas and any extra whitespace
        extracted['statutes_split'][i][j] = extracted['statutes_split'][i][j].replace(',', '').strip()
        extracted['statutes_split'][i][j] = extracted['statutes_split'][i][j].replace('  ', ' ')
        
```

```python
def matrix_dot_vector(a:list[list[int|float]],b:list[int|float])-> list[int|float]:
    if len(a[0]) != len(b):
        return -1
    c = []
    for i in a:
        sum = 0
        for j in range(len(b)):
            sum += (i[j]*b[j])
        c.append(sum)
    return c


a = [[1,2],[2,4]]
b = [1,2]
matrix_dot_vector(a,b)
```

```python
def transpose_matrix(a: list[list[int|float]]) -> list[list[int|float]]:
    b = []
    for i in range(len(a[0])):
      b.append([a[i],b[i]])
    return b

a = [[1,2],[3,4],[5,6]]

transpose_matrix(a)
```
